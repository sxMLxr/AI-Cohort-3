```python
#!pip install pandas matplotlib seaborn

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data
data = '''
Date,Increased,Present level,Decreased
Jun 24 1965,7,39,33
Mar 25 1977,7,37,42
Jun 19 1986,7,35,49
Jul 9 1993,6,27,65
Jun 5 1995,7,24,65
Jul 7 1995,7,27,62
Feb 26 1999,10,41,44
Sep 11 2000,13,41,38
Mar 26 2001,10,41,43
Jun 11 2001,14,42,41
Oct 19 2001,8,30,58
Jun 3 2002,12,36,49
Sep 2 2002,17,26,54
Jun 12 2003,13,37,47
Jun 9 2004,14,33,49
Jun 6 2005,16,34,46
Dec 9 2005,15,31,51
Apr 7 2006,15,35,47
Jun 8 2006,17,42,39
Jun 4 2007,16,35,45
Jun 5 2008,18,39,39
Jul 10 2009,14,32,50
Jul 8 2010,17,34,45
Jun 9 2011,18,35,43
Jun 7 2012,21,42,35
Jun 13 2013,23,40,35
Feb 6 2014,27,35,36
Jun 5 2014,22,33,41
Jun 15 2015,25,40,34
Jun 7 2016,21,38,38
Jun 7 2017,24,38,35
Jun 1 2018,28,39,29
Jan 21 2019,30,37,31
Jun 3 2019,27,37,35
May 28 2020,34,36,28
Jun 1 2021,33,35,31
Jul 5 2022,27,31,38
'''

# Load data into a pandas DataFrame
from io import StringIO
df = pd.read_csv(StringIO(data), parse_dates=['Date'])

# Set seaborn style
sns.set()

# Create a line chart using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Increased', label='Increased')
sns.lineplot(data=df, x='Date', y='Present level', label='Present level')
sns.lineplot(data=df, x='Date', y='Decreased', label='Decreased')

# Add vertical lines for each year
years = np.arange(df['Date'].dt.year.min(), df['Date'].dt.year.max() + 1)
for year in years:
    plt.axvline(pd.to_datetime(str(year)), color='gray', linestyle='--', linewidth=0.5)

# Add horizontal lines for each percentage point
percentage_points = np.arange(0, 101, 1)
for perc in percentage_points:
    plt.axhline(perc, color='gray', linestyle='--', linewidth=0.5, alpha=0.25)

# Add title and labels
plt.title("America's Views on Immigration (1965-2022)")
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend()

# Show the plot
plt.show()

```
